```python
import pandas as pd
import datetime
import dtale as dt
import pandas_profiling
from sklearn.preprocessing import (MinMaxScaler,OneHotEncoder)
```

```python
#CITIZEN_ID turn to index
#Combine Maried in Army with normal Married
#Combine Wife and Husband for Spouse
```

```python
raw_data = pd.read_excel("Train.xlsx")
data_prepro = raw_data.copy()
```

```python
data_prepro.info()
```

```python
data_prepro['Birthday']=data_prepro['Birthday'].map(lambda x:2044 - int(x[-4:]))
```

```python
data_prepro = data_prepro.drop((data_prepro.loc[(data_prepro['Base Area'] == '?') |
                  (data_prepro['Role'] == '?') |
                  (data_prepro['Employment Sector'] == '?') |
                  (data_prepro['Employment Sector'] == 'Unemployed')]).index, axis = 0)
```

```python
data_prepro = data_prepro.set_index('CITIZEN_ID')
```

```python
data_prepro = data_prepro.drop(['Name','Education Level'],axis=1)
```

```python
data_prepro.columns = ['Age', 'Native Continent',
       'Marital Status', 'Lives with', 'Base Area',
       'Years of Education', 'Employment Sector', 'Role',
       'Working Hours per week', 'Money Received', 'Ticket Price', 'Income']
```

```python
dt.show(data_prepro)
```

```python
BA_freq_map = data_prepro['Base Area'].value_counts().to_dict()
data_prepro['Base Area'] =data_prepro['Base Area'].map(BA_freq_map)
```

```python
data_prepro['Marital Status'] = data_prepro['Marital Status'].map(lambda x: x if x != 'Married - Spouse in the Army' else 'Married')
```

```python
spouse = ['Wife','Husband']
data_prepro['Lives with'] = data_prepro['Lives with'].map(lambda x: 'Spouse' if x in spouse else x)
```

```python
sectors = ['Private','Public ']
data_prepro['Employment Sector'] = data_prepro['Employment Sector'].map(lambda x: strip(x[0:7]) if x[0:7] in sectors else x)
```
